```python
# Kameo investeringer – Analyse
# Denne notebooken parser investeringene mine fra Kameo, og gir innsikt i:
# - Investeringer og renteinntekter per låntaker
# - Totale investeringer
# - Forventede renteinntekter fremover
```

```python
import pandas as pd
import re
from pathlib import Path
from datetime import datetime
import matplotlib.pyplot as plt
import plotly.express as px
import plotly.graph_objects as go
```

```python
# Parse functions

def load_investment_file(path):
    with open(path, encoding="utf-8") as f:
        content = f.read()
    loans = re.split(r'\n(?=\w.+ - \d+ \| Løpetid)', content)
    return loans

def parse_loan_block(block):
    lines = block.strip().split('\n')
    header = lines[0]
    match = re.match(r"(.*?) - (\d+)\s+\| Løpetid: (\d+)[^|]+\| Rente: ([\d,]+)%", header)
    if not match:
        return None
    company, loan_id, duration, interest = match.groups()
    return {
        "company": company.strip(),
        "loan_id": int(loan_id),
        "duration_months": int(duration),
        "interest_rate": float(interest.replace(',', '.'))
    }

def parse_transactions(block, loan_id):
    lines = block.strip().split('\n')
    data_lines = []
    in_data = False
    for line in lines:
        if line.startswith("Dato"):
            in_data = True
            continue
        if in_data:
            if line.startswith("Totale renteinntekter"):
                break
            parts = line.split('\t')
            if len(parts) == 6:
                date_str, txn_type, amount, _, _, amount_nok = parts
                try:
                    data_lines.append({
                        "loan_id": loan_id,
                        "date": pd.to_datetime(date_str, format="%Y-%m-%d"),
                        "transaction_type": txn_type.strip(),
                        "amount_nok": float(amount_nok.replace(",", ".").replace("−", "-").replace(" ", ""))
                    })
                except Exception as e:
                    print(f"Feil i linje: {line}\n{e}")
    return data_lines
```

```python
# Load, and parse data

loans_raw = load_investment_file("investments.txt")

parsed_loans = []
parsed_transactions = []

for block in loans_raw:
    loan = parse_loan_block(block)
    if loan:
        parsed_loans.append(loan)
        parsed_transactions.extend(parse_transactions(block, loan["loan_id"]))

df_loans = pd.DataFrame(parsed_loans)
df_txns = pd.DataFrame(parsed_transactions)
df_txns_merged = df_txns.merge(df_loans, on="loan_id")
```

```python
# Monthly interest

df_txns_merged["year_month"] = df_txns_merged["date"].dt.to_period("M")
monthly_interest = df_txns_merged[df_txns_merged["transaction_type"] == "Renteinntekt"]
monthly_sum = monthly_interest.groupby("year_month")["amount_nok"].sum()

fig = px.bar(
    monthly_sum.reset_index(),
    x="year_month",
    y="amount_nok",
    title="Renteinntekter per måned",
    labels={"year_month": "Måned", "amount_nok": "Sum rente (NOK)"},
    text_auto=True
)
fig.update_layout(xaxis_tickangle=-45)
fig.show()
```

```python
# Aggregering per låntaker

# 1. Summer alle transaksjoner per låntaker og transaksjonstype
summary = df_txns_merged.groupby(["company", "transaction_type"])["amount_nok"].sum().unstack(fill_value=0)

# 2. Sikre at alle nødvendige kolonner finnes
for col in ["Tildeling", "Tilbakebetaling", "Renteinntekt", "Forsinkelsesrente"]:
    if col not in summary.columns:
        summary[col] = 0.0

# 3. Legg til aktiv kapital og netto rente
summary["aktiv_kapital"] = summary["Tildeling"] + summary["Tilbakebetaling"]
summary["netto_rente"] = summary["Renteinntekt"] + summary["Forsinkelsesrente"]

# 4. Marker hvilke lån som er aktive basert på sum(Tildeling + Tilbakebetaling)
df_loans["aktiv_kapital"] = df_loans["loan_id"].map(
    df_txns_merged[df_txns_merged["transaction_type"].isin(["Tildeling", "Tilbakebetaling"])]
    .groupby("loan_id")["amount_nok"].sum()
)

df_loans["aktiv"] = df_loans["aktiv_kapital"] != 0

# 5. Tell antall lån og aktive lån per låntaker
loan_counts = df_loans.groupby("company")["loan_id"].count().rename("antall_lån")
active_counts = df_loans[df_loans["aktiv"]].groupby("company")["loan_id"].count().rename("aktive_lån")

counts_df = pd.concat([loan_counts, active_counts], axis=1).fillna(0).astype(int)

# 6. Slå sammen med summary-tabellen
summary = summary.merge(counts_df, left_index=True, right_index=True, how="left")

# 7. Reorder kolonner
summary = summary[[
    "antall_lån", "aktive_lån",
    "Tildeling", "Tilbakebetaling", "aktiv_kapital",
    "Renteinntekt", "Forsinkelsesrente", "netto_rente"
]]

# 8. Legg til totalrad
summary.loc["Totalt"] = summary.sum(numeric_only=True)

# 9. Sorter etter netto_rente
summary = summary.sort_values("netto_rente", ascending=False)

# 10. Vis resultat
fig = go.Figure(data=[go.Table(
    header=dict(
        values=["<b>" + col + "</b>" for col in summary.reset_index().columns],
        fill_color="lightblue",
        align="left"
    ),
    cells=dict(
        values=[summary.reset_index()[col] for col in summary.reset_index().columns],
        fill_color="lavender",
        align="left"
    )
)])
fig.update_layout(title="Sammendrag per låntaker (interaktiv tabell)", height=600)
fig.show()
```

```python
# Future interest (prognosis)

today = pd.Timestamp.today()

# Sjekk at "year_month"-kolonnen eksisterer
if "year_month" not in df_txns_merged.columns:
    df_txns_merged["year_month"] = df_txns_merged["date"].dt.to_period("M")

future = df_txns_merged[
    (df_txns_merged["transaction_type"] == "Renteinntekt") &
    (df_txns_merged["date"] > today)
]

if not future.empty:
    future.groupby("year_month")["amount_nok"].sum().plot(kind="bar", title="Fremtidige renteinntekter", figsize=(10, 4))
    plt.ylabel("Rente (NOK)")
    plt.xlabel("Måned")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
else:
    print("Ingen fremtidige renteinntekter registrert etter i dag:", today.strftime("%Y-%m-%d"))
```
